```python
from langchain_community.chat_models import ChatCohere
import logging
from langchain.schema import HumanMessage, AIMessage

# Set up logging
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

def initialize_cohere_chat(api_key):
    """
    Initialize and test Cohere chat model with proper error handling
    """
    logger.info("Loading Cohere chat pipeline...")
    
    try:
        # Initialize the chat model with temperature and other parameters
        cohere_chat_model = ChatCohere(
            cohere_api_key=api_key,
            model="command",  # Explicitly specify the model
            temperature=0.7,
            max_tokens=2048
        )
        
        # Test the model with a simple message
        test_message = HumanMessage(content="Hello! How are you?")
        logger.info("Sending test message to Cohere API...")
        
        # Remove the callbacks parameter that was causing the error
        test_response = cohere_chat_model.invoke([test_message])
        
        logger.info("Successfully received response from Cohere")
        return cohere_chat_model
        
    except ValueError as ve:
        logger.error(f"Invalid API key or configuration: {str(ve)}")
        raise
    except ConnectionError as ce:
        logger.error(f"Connection error when trying to reach Cohere API: {str(ce)}")
        raise
    except Exception as e:
        logger.error(f"Unexpected error initializing Cohere chat: {str(e)}")
        raise

if __name__ == "__main__":
    # Your API key
    cohere_key = "n7OYjtvEIrpQ1q03WGhuypIktuhhHgfoKAvjWGgX"
    
    try:
        chat_model = initialize_cohere_chat(cohere_key)
        print("Cohere chat model initialized successfully!")
        
    except Exception as e:
        print(f"Failed to initialize Cohere chat model: {str(e)}")
```
