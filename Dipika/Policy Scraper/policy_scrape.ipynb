```python
import requests
from bs4 import BeautifulSoup
```

```python

# Send HTTP request
url = "https://bitpay.com/about/privacy-eu/"
response = requests.get(url)

# Check if successful
if response.status_code == 200:
    # Parse HTML content
    soup = BeautifulSoup(response.content, 'html.parser')

    # Remove unnecessary tags
    for script in soup(["script", "style"]):
        script.decompose()

    # Get text
    text = soup.get_text()

    # Break into lines and remove leading/trailing space
    lines = (line.strip() for line in text.splitlines())

    # Break multi-headlines into a line each
    chunks = (phrase.strip() for line in lines for phrase in line.split("  "))

    # Drop blank lines
    text = '\n'.join(chunk for chunk in chunks if chunk)

    # Save text to file
    with open("BitPay_policy.txt", 'w', encoding='utf-8') as file:
        file.write(text)

else:
    print("Failed to retrieve page")
```

```python
# Print BitPay privacy policy text
with open("BitPay_policy.txt", "r") as file:
    print("BitPay Policy:")
    print(file.read())
```

```python
# URL of the privacy policy page
url = "https://www.moonpay.com/legal/privacy_policy"

# Send a GET request to the page
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the page content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Remove unnecessary tags such as script and style
    for script in soup(["script", "style"]):
        script.decompose()

    # Find the main content of the privacy policy (adjust this class or tag to match the website's structure)
    main_content = soup.find('article')  # Adjust this based on the page structure

    if main_content:
        # Get all text within the content
        text = main_content.get_text(separator='\n', strip=True)
    else:
        # Fallback to getting all text if the main content is not found
        text = soup.get_text(separator='\n', strip=True)

    # Save the extracted text to a file
    with open("MoonPay_policy.txt", 'w', encoding='utf-8') as file:
        file.write(text)

    print("Policy saved to MoonPay_policy.txt")

else:
    print(f"Failed to retrieve page, status code: {response.status_code}")
```

```python
# Print MoonPay privacy policy text
with open("MoonPay_policy.txt", "r") as file:
    print("MoonPay Policy:")
    print(file.read())
```
